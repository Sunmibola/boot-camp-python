```python
# We first have to import pandas and numpy using import statement in python
import pandas as pd 
import numpy as np
```

```python
#we have to import the csv file to be analysed by  uploading into the files
#create a data frame that can be read in the csv format
df=pd.DataFrame(pd.read_csv('//content/train (1).csv'))
#then print the first 5 part of the data set to catch a glimpse 
df.head
```

```python
#to determine the dimension of the data set 
df.shape
```

```python
#to count the total number of NaN values in the dataframe 
df.isnull().sum()
```

```python
#from the information above , 
#cabin NaN value is over 40% of the total rows 
# Age NaN value is  over 15% of the rows. we need to drop the largest NaN value 
drop_col =df.isnull().sum()[df.isnull().sum()>(40/100 *df.shape[0])]
drop_col
```

```python
#to determine the index of the dropped column and drop
drop_col.index
```

```python
#to see changes made and sum of NaN left 

df.isnull().sum()
```

```python
#to fill in NaN  in Age with values , 
#lets insert the mean of the age 
df.fillna(df.mean(),inplace=True)
df.isnull().sum()
```

```python
#we still have 2 NaN in embarked, 
#decribe embark to see detail 
df['Embarked'].describe()
```

```python
#most value in 'embarked' is in S, thus lets insertn s into the 2 left NaN
df['Embarked'].fillna ('S',inplace= True)
```

```python
#correlation analysis to show the relationsip between the variables 
#to show how each variable move with the other 
df.corr
```

```python
#lets create a new variable by merging the two 'parch' and Sibsp
df['Familysize'] = df['SibSp'] + df['Parch']
```

```python
#Lets drop the two variables since a new one that can represent them had been created
df.drop(['SibSp', 'Parch'], axis =1 , inplace=True)
```

```python
#correlate family size against surival rate
df['Familysize'].corr(df['Survived']) 
#below shows that there is a positive relationship between Family size and survival rate
```

```python
#correlate Age against surival rate
df['Age'].corr(df['Survived']) 
#below shows that there is a negative relationship between Age and Survival rate
```

```python
#correlate pclass against surival rate
df['Pclass'].corr(df['Survived']) 
#below shows that there is a negative relationship between Pclass and Survival rate
```

```python
#to check if being alone will affect survival rate
df['Alone']=[0 if df['Familysize'] [i]>0 else 1 for i in df.index]
df.head(15)
```

```python
#to calculate the mean of those who were alone '0' and survived 
#and those who are with their family "1'and survive
df.groupby(['Alone'])['Survived'].mean()
```

```python
df['Alone'].corr(df['Survived'])
#there is a negative relationship between Being alone and the survival rate 
```

```python
#the correlation between being alone and fares shows a  negatiive relationship
df['Alone'].corr (df['Fare'])
```

```python
#to assign values 1 for female and 0 to male and to filter out the survival rate for each
df['Sex']=[0 if df['Sex'] [i]=='male' else 1 for i in df.index] #  1 for female , 0 for male
df.groupby(['Sex'])['Survived'].mean()
```

```python
#to determine the survival relationship with the embarking station
df.groupby(['Embarked'])['Survived'].mean()
```
